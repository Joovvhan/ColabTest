```
import numpy as np

x = np.random.choice(range(0, 1000), 50, replace=False)
y = np.random.choice(range(1000, 2000), 50, replace=False)
z = np.random.choice(range(2000, 3000), 50, replace=False)

print(x)

print(y)

print(z)

A = [(x[i], y[i], z[i]) for i in range(50)]

print(A)

sorted(A, key=lambda index: index[0])
```
