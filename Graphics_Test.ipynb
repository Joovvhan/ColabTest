```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

plt.gca().get_xaxis().set_visible(False)
plt.gca().get_yaxis().set_visible(False)
plt.gca().set_facecolor('w')

plt.plot(x,mlab.normpdf(x, mu, sigma), color = 'k')
plt.plot(x, -0.01 * np.ones(len(x)), color = 'k', linestyle='dashed')
plt.show()
```

```
def plot_cube(cube_definition, ax, corner):
    cube_definition_array = [
        np.array(list(item))
        for item in cube_definition
    ]

    points = []
    points += cube_definition_array
    vectors = [
        cube_definition_array[1] - cube_definition_array[0],
        cube_definition_array[2] - cube_definition_array[0],
        cube_definition_array[3] - cube_definition_array[0]
    ]

    points += [cube_definition_array[0] + vectors[0] + vectors[1]]
    points += [cube_definition_array[0] + vectors[0] + vectors[2]]
    points += [cube_definition_array[0] + vectors[1] + vectors[2]]
    points += [cube_definition_array[0] + vectors[0] + vectors[1] + vectors[2]]

    points = np.array(points)

    edges = [
        [points[0], points[3], points[5], points[1]],
        [points[1], points[5], points[7], points[4]],
        [points[4], points[2], points[6], points[7]],
        [points[2], points[6], points[3], points[0]],
        [points[0], points[2], points[4], points[1]],
        [points[3], points[6], points[7], points[5]]
    ]

    faces = Poly3DCollection(edges, linewidths=0.05, edgecolors='k', alpha = 1 - corner / 5)
    faces.set_facecolor((1, 1 - corner/10, 1 - corner/10, 0))

    ax.add_collection3d(faces)
    
    
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

cubeList = list()
xyz = list()

for x in range(5):
    for y in range(5):
        for z in range(5):
            cubeList.append([(x, y, z), (x + 1, y, z), (x, y + 1, z), (x, y, z + 1)])
            xyz.append(min([x, y, z]) + 1)

for i in range(len(cubeList)):
    plot_cube(cubeList[i], ax, xyz[i])

# Plot the points themselves to force the scaling of the axes

ax.set_xlim((0, 6))
ax.set_ylim((6, 0))
ax.set_zlim((0, 6))

ax.set_xticks([0, 1, 2, 3, 4, 5])                               
ax.set_yticks([1, 2, 3, 4, 5])                               
ax.set_zticks([0, 1, 2, 3, 4, 5])

ax.set_aspect('equal')

    
ax.view_init(elev=30, azim=-60)
    
# ax.set_aspect('equal')

plt.xticks(fontsize=24)
plt.xlabel('Pitch', fontsize=24, labelpad = 24)

plt.yticks(fontsize=24)
plt.ylabel('Freqeuncy', fontsize=24, labelpad = 24)

# ax.set_zticks(fontsize=24)
ax.set_zlabel('Power', fontsize=24, labelpad = 12)

ax.set_zticklabels([0, 1, 2, 3, 4, 5], fontsize=24)
plt.show()
```

```
def plot_cube(cube_definition, ax, cFactor, aFactor):
    cube_definition_array = [
        np.array(list(item))
        for item in cube_definition
    ]

    points = []
    points += cube_definition_array
    vectors = [
        cube_definition_array[1] - cube_definition_array[0],
        cube_definition_array[2] - cube_definition_array[0],
        cube_definition_array[3] - cube_definition_array[0]
    ]

    points += [cube_definition_array[0] + vectors[0] + vectors[1]]
    points += [cube_definition_array[0] + vectors[0] + vectors[2]]
    points += [cube_definition_array[0] + vectors[1] + vectors[2]]
    points += [cube_definition_array[0] + vectors[0] + vectors[1] + vectors[2]]

    points = np.array(points)

    edges = [
        [points[0], points[3], points[5], points[1]],
        [points[1], points[5], points[7], points[4]],
        [points[4], points[2], points[6], points[7]],
        [points[2], points[6], points[3], points[0]],
        [points[0], points[2], points[4], points[1]],
        [points[3], points[6], points[7], points[5]]
    ]

    faces = Poly3DCollection(edges, linewidths=0.05, edgecolors='k', alpha = 1 - aFactor / 5)
    faces.set_facecolor((0.8 * cFactor/15, 0.5 * (1 - cFactor/15), 1 - cFactor/15, 0))

    ax.add_collection3d(faces)
    
    
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

cubeList = list()
aList = list()
cList = list()

for x in range(5):
    for y in range(5 - x):
        for z in range(5 - x - y):
            cubeList.append([(x, y, z), (x + 1, y, z), (x, y + 1, z), (x, y, z + 1)])
            aList.append(min([x, y, z]) + 1)
            cList.append(x + y + z + 1)

for i in range(len(cubeList)):
    plot_cube(cubeList[i], ax, cList[i], aList[i])

# Plot the points themselves to force the scaling of the axes

ax.set_xlim((0, 6))
ax.set_ylim((6, 0))
ax.set_zlim((0, 6))

ax.set_xticks([0, 1, 2, 3, 4, 5])                               
ax.set_yticks([1, 2, 3, 4, 5])                               
ax.set_zticks([0, 1, 2, 3, 4, 5])

ax.set_aspect('equal')

    
ax.view_init(elev=30, azim=-60)
    
# ax.set_aspect('equal')


plt.show()
```
